```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data2018.csv')
data.head()
```

```python
print(data.columns)
```

```python
data = data.iloc[:,[1,5,14,16,22,24,25]]
data.head()
```

```python
[row, col] = data.shape
for i in range(row):
    data.loc[i,'POS'] = ((data.loc[i,'FGA']+data.loc[i,'TOV'])-(data.loc[i,'ORB']+(0.4*data.loc[i,'FTA'])))
data.head()
```

```python
data = data.iloc[:-1,:]
data
```

```python
[row, col] = data.shape
for i in range(row):
    data.loc[i,'OER'] = (data.loc[i,'PTS']/data.loc[i,'POS'])*100
    data.loc[i,'DER'] = (data.loc[i,'OPTS']/data.loc[i,'POS'])*100
```

```python
oeravg = data.loc[:,'OER'].mean(axis=0)
deravg = data.loc[:,'DER'].mean(axis=0)
print(oeravg)
print(deravg)
```

```python
for i in range(row):
    data.loc[i,'Odiff'] = data.loc[i,'OER']-oeravg
    data.loc[i,'Ddiff'] = data.loc[i,'DER']-deravg
    data.loc[i,'Tempo'] = data.loc[i,'POS']/82
```

```python
tempoavg = data.loc[:,'Tempo'].mean(axis=0)
print(tempoavg)
```

```python
for i in range(row):
    data.loc[i,'Tdiff'] = data.loc[i,'Tempo']-tempoavg
data
```

```python
def GameSim(team1, team2):
    team1odiff = np.random.normal(data.loc[data['Team']==team1,'OER'],scale=5)-oeravg
    team1ddiff = np.random.normal(data.loc[data['Team']==team1,'DER'],scale=5)-deravg
    team2odiff = np.random.normal(data.loc[data['Team']==team2,'OER'],scale=5)-oeravg
    team2ddiff = np.random.normal(data.loc[data['Team']==team2,'DER'],scale=5)-deravg
    team1expeff = (oeravg + team1odiff + team2ddiff)*(1.02)
    team2expeff = (oeravg + team2odiff + team1ddiff)*(0.98)
    matchtempo = data.loc[data['Team']==team1,'Tdiff'].to_numpy()+data.loc[data['Team']==team2,'Tdiff'].to_numpy()+tempoavg
    team1score = (team1expeff/100)*matchtempo
    team2score = (team2expeff/100)*matchtempo
    score = [team1score, team2score]
    return score
```

```python
def simulator(team1, team2):
    team1win = 0
    team2win = 0
    team1score = 0
    team2score = 0
    matchcount=0
    for i in range(10000):
        gm = GameSim(team1,team2)
        if gm[0] > gm[1]:
            team1win+=1
        elif gm[0]<gm[1]:
            team2win+=1
        else:
            matchcount-=1
        team1score+=gm[0]
        team2score+=gm[1]
        matchcount+=1
    #if team1win>team2win:
     #   return 1
    #elif team1win<team2win:
     #   return 0
    print(team1, 'chances of winning = ',(team1win/matchcount)*100,'%')
    print(team2, 'chances of winning = ',(team2win/matchcount)*100,'%')
    print(team1, 'possible score = ',(round((team1score/matchcount)[0])))
    print(team2, 'possible score = ',(round((team2score/matchcount)[0])))
```

```python
simulator('New York Knicks', 'Golden State Warriors')
```

```python
[rowm,colm] = match.shape
for i in range(rowm):
    if match.iloc[i,4]>match.iloc[i,2]:
        match.loc[i,'Winner'] = match.iloc[i,3]
    elif match.iloc[i,4]<match.iloc[i,2]:
        match.loc[i,'Winner'] = match.iloc[i,1]
match.head()
```

```python
count=0
for i in range(rowm):
    game = simulator(match.loc[i,'HomeTeam'],match.loc[i,'AwayTeam'])
    if game == 1:
        match.loc[i,'Predicted'] = match.loc[i,'HomeTeam']
    elif game ==0:
        match.loc[i,'Predicted'] = match.loc[i,'AwayTeam']
    if match.loc[i,'Winner'] == match.loc[i,'Predicted']:
        count+=1
print((count/rowm)*100)
```
